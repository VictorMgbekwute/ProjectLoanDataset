## Investigation Overview

In this Investigation I wanted to look for certain features of the dataset that can help predict if OnTimeProsperPayment from the customers. I'm interested in analysing the MonthlyLoanPayment, EmploymentStatusDuration,OnTimeProsperPayment and ProsperRating(Alpha). 


## Dataset Overview

There are 113937 loans in the Dataset with 81 features. 59 columns of the variables are numeric in nature while 22 are categorical variables.

````python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
````

````python
# load in the dataset into a pandas dataframe
prosper_loan = pd.read_csv('prosperLoanData.csv')
````

## (Distribution of OnTimeProsperPayments)


A look into The distribution of the OnTimeProsperPayment showed a right skewed distribution with long tail on a standard scale.

````python
bin_edges = np.arange(0,prosper_loan['OnTimeProsperPayments'].max()+10,10)
sb.histplot(prosper_loan['OnTimeProsperPayments'], bins = bin_edges, kde = False,)
plt.xlabel('OnTimeProsperPayments')
plt.ylabel('Number of OnTimePayment made')
plt.title('OnTimeProsperPayments on Standard scale');
````

## (LoanOriginalAmount Vs MonthlyLoanPayment)


The distribution showing a positive correlation between these two features. It showed the highest LoanOriginalAmount of 
35000 has a MonthlyLoanPayment between 1000.

````python
plt.figure(figsize = [8,8])
plt.scatter(data = prosper_loan, x = 'MonthlyLoanPayment', y = 'LoanOriginalAmount', alpha = 1/10)
plt.xlabel('MonthlyLoanPayment ($)')
plt.ylabel('LoanOriginalAmount ($)')
plt.title('LoanOriginalAmount and MonthlyLoanPayment Distribution');
````

## (MonthlyLoanPayment vs OnTimeProsperPayments vs ProsperScore)



The distribution showed a positive corellation between OnTimeProsperPayment and MonthlyLoanPayment as the ProsperScore increases.

````python
plt.scatter(data = prosper_loan, x = 'LoanOriginalAmount', y = 'MonthlyLoanPayment', c = 'ProsperScore', cmap = 'inferno_r')
plt.colorbar(label = 'ProsperScore')
plt.xlabel('OnTimeProsperPayments')
plt.ylabel('MonthlyLoanPayment')
plt.title('MonthlyLoanPayment,OnTimeProsperPayments and ProsperScore');
````
